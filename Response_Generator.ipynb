```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def generate_reply(email_text, predicted_category):
    """
    Generate a context-aware reply based on email text and category.
    """
    # Dynamic prompt templates
    if predicted_category == "IT":
        prompt = (
            "You are an IT support agent. Write a polite and helpful reply to this email:\n"
            f"\"{email_text}\"\n\nResponse:"
        )
    elif predicted_category == "HR":
        prompt = (
            "You are an HR representative. Write a professional reply to this email:\n"
            f"\"{email_text}\"\n\nResponse:"
        )
    else:
        prompt = (
            "Write a polite and professional reply to this email:\n"
            f"\"{email_text}\"\n\nResponse:"
        )

    # Generate response
    generated = generator(
        prompt,
        max_length=200,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.7
    )

    reply = generated[0]["generated_text"].replace(prompt, "").strip()

    return {"response": reply}
```

```python
input_data = {
    "email_text": "Hi, I forgot my laptop password. Please help.",
    "predicted_category": "IT"
}

result = generate_reply(input_data["email_text"], input_data["predicted_category"])
print(result)
```
